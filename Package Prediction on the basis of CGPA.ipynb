```python
placement_data = pd.read_csv('/content/Placement.csv')
print(placement_data)
```

```python
x = placement_data['cgpa']
# show the separated attribute/feature I.E CGPA
print(x)
```

```python
y = placement_data['package']
print(y)
```

```python
[x_train, x_test, y_train, y_test] = train_test_split(x,y, test_size = 0.25, random_state = 2)
```

```python
print('The size of x_train is', x_train.shape)
print('The size of y_train is', y_train.shape)
print('The size of x_test is', x_test.shape)
print('The size of y_test is', y_test.shape)
```

```python
x_train_2d = x_train.values.reshape(-1,1)
x_test_2d = x_test.values.reshape(-1,1)
print('The size of training data is', x_train_2d.shape)
print('The size of test data is', x_test_2d.shape)
```

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(x_train_2d, y_train)
```

```python
#see the parameters
print('The coefficient is', round(linear_model.coef_[0],2))
print('The intercept is' , round(linear_model.intercept_,2))
```

```python
#make prediction on the training dataset
x_train_data_pred = linear_model.predict(x_train_2d)
```

```python
#plot the model
plt.scatter(x_train,y_train)
plt.plot(x_train,x_train_data_pred,color='r')
plt.show()
```

```python
#make predictions on the test set
x_test_data_pred = linear_model.predict(x_test_2d)
#print(x_test_data_pred)
```

```python
#model evalutaiton on training data and test data
r_square_train = round(r2_score(y_true = y_train, y_pred = x_train_data_pred),2)
r_square_test = round(r2_score(y_true = y_test, y_pred = x_test_data_pred))
```

```python
#evalute the model
print('The R-squared for the training set is', r_square_train)
print('The R-squared for the test set is',r_square_test)
```

```python
#plot the model with the test set
plt.scatter(x_train,y_train)
plt.scatter(x_test,y_test)
plt.plot(x_test,x_test_data_pred, color='g')
plt.show()
```

```python
#make predictive model
#Now prdict on new input unseen data points
input_data = (6.93)
input_data_as_numpy_array = np.asarray(input_data)
#Reshape the numpy array as we are predicting only for single data intance at a time
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = linear_model.predict(input_data_reshaped)
print(prediction)
```
